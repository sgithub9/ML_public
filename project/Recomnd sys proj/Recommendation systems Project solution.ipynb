```python
from surprise import KNNWithMeans
from surprise import accuracy
from surprise.model_selection import train_test_split
from surprise import Reader
from surprise.model_selection import KFold
from surprise import Dataset
import pandas as pd
```

```python
data = pd.read_csv('ratings_Electronics.csv',header = None)
```

```python
data.head()
```

```python
df = data.drop([3], axis=1) # dropping the last column of timestamp values since they dont seem to be relevant
```

```python
df.columns = ['userId', 'productId','rating'] #renaming columns
```

```python
df.head()
```

```python
df_new = df.groupby('userId').filter(lambda x: len(x)>50)
df_new.describe()
```

```python
df_new.head()
```

```python
df_new.dtypes
```

```python
df_new.hist(layout=(1,2))
```

```python
#We could see that most (70,000 out of 1.2 lakhs) of the ratings fall into '5' rating category.
```

```python
reader = Reader(rating_scale=(1, 5))
```

```python
dataset = Dataset.load_from_df(df_new, reader)
```

```python
trainset, testset = train_test_split(dataset, test_size=0.3)
```

```python
# define a cross-validation iterator
kf = KFold(n_splits=3)
```

```python
#'user_based': True indicating that this would be a user-user collab filtering
#This is because the objective is to "Build a recommendation system to recommend products to
#customers based on the their previous ratings for other products", so similar ratings by other users would help
algo = KNNWithMeans(k=50, sim_options={'name': 'pearson_baseline', 'user_based': True})
algo.fit(trainset)
```

```python
# run the trained model against the testset
test_pred = algo.test(testset)
```

```python
test_pred
```

```python
# get RMSE
print("User-based Model : Test Set")
accuracy.rmse(test_pred, verbose=True)
```

```python
for trainset, testset in kf.split(dataset):
    # train and test algorithm.
    algo = KNNWithMeans(k=50, sim_options={'name': 'pearson_baseline', 'user_based': True})
    algo.fit(trainset)
    # run the trained model against the testset
    test_pred = algo.test(testset)
    print("User-based Model : Test Set")
    accuracy.rmse(test_pred, verbose=True)

    
```

```python
#From the above its clear that RMSE isnt falling below the level of 1.051, so we could try changing the hyper parameter 'k'
```

```python
algo = KNNWithMeans(k=1000, sim_options={'name': 'pearson_baseline', 'user_based': True})
algo.fit(trainset)
```

```python
test_pred = algo.test(testset)
print("User-based Model : Test Set")
accuracy.rmse(test_pred, verbose=True)
```

```python
#Even changing the 'K' hyper param doesnt reduce the RMSE score below 1. And as we also see from the test prediction
# the model is giving wrong predictions since for uid='A2ED50E3KWKUKW' and iid='B005IMU06O' which is an existing entry
# it shows as though it doesnt exist by giving the output for parameter "details={'was_impossible': True"

#So we could abandon the 'pearson_baseline' simulation technique and try another 'cosine'
```

```python
sim_options = {
    "name": "cosine",
    "user_based": True,  # Compute  similarities between user choices
}
algo = KNNWithMeans(k=50, sim_options=sim_options)
algo.fit(trainset)
```

```python
test_pred = algo.test(testset)
print("User-based Model : Test Set")
accuracy.rmse(test_pred, verbose=True)
```

```python
#The RMSE is even worse so we abandon cosine technique as well
```

```python
from surprise import SVD
from surprise.model_selection import GridSearchCV
param_grid = {
    "n_epochs": [5, 10],
    "lr_all": [0.002, 0.005],
    "reg_all": [0.4, 0.6]
}
gs = GridSearchCV(SVD, param_grid, measures=["rmse", "mae"], cv=3)

gs.fit(dataset)

print(gs.best_score["rmse"])
print(gs.best_params["rmse"])
```

```python
#Seems like SVD using GrdiSearchCV gives some improved results for RMSE
algo = gs.best_estimator['rmse']#getting the best estimator
algo.fit(trainset)
```

```python
#Randomly chosen userid and productid
uid='A2VALCFTOWFYXZ'
iid='B00558OP3U'
prediction = algo.predict(uid, iid)
```

```python
df_grouped_by_pdt_rating_mean = pd.DataFrame(df_new.groupby('productId')['rating'].mean())
```

```python
df_grouped_by_pdt_rating_mean['rating_count'] = pd.DataFrame(df_new.groupby('productId')['rating'].count())
```

```python
df_grouped_by_pdt_rating_mean = df_grouped_by_pdt_rating_mean.reset_index()
```

```python
df_grouped_by_pdt_rating_mean.info()
```

```python
df_grouped_by_pdt_rating_mean.head()
```

```python
#Creating a df with products of rating>3 and with >20 ratings, ones which could be considered popular
popular_products = df_grouped_by_pdt_rating_mean.loc[df_grouped_by_pdt_rating_mean['rating']>3]
popular_products = popular_products.loc[popular_products['rating_count']>20]#filering for items with atleast 21 or more ratings
popular_products = popular_products.reset_index()
popular_products.sort_values(by='rating_count', ascending=False).head()
```

```python
df_pdts_of_user = df_new.loc[df_new['userId']=='A3QKIR8J56EUI1',['productId','rating']].reset_index()
keys = list(df_pdts_of_user.columns.values)
i1 = popular_products.set_index(keys).index
i2 = df_pdts_of_user.set_index(keys).index
popular_products[~i1.isin(i2)].sort_values(by='rating_count', ascending=False).head()
```

```python
#The above output shows the results of the popular products recommendation for userid - 'A3QKIR8J56EUI1'.
#It can be encapsulated inside a method as shown below
def recommendProductsToUser(userId, df_orig, df_popular_products):
    df_pdts_of_user = df_orig.loc[df_orig['userId']==userId,['productId','rating']].reset_index()
    keys = list(df_pdts_of_user.columns.values)
    i1 = popular_products.set_index(keys).index
    i2 = df_pdts_of_user.set_index(keys).index
    return df_popular_products[~i1.isin(i2)].sort_values(by='rating_count', ascending=False).head()
```

```python
recommendProductsToUser('A3QKIR8J56EUI1', df_new, popular_products)
```
